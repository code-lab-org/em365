Import dependent libraries: pandas (`pd`) for data manipulation, statsmodels (`sm`) for statistics, statsmodels formulas (`smf`) for building models using formulas, statsmodels graphics (`smg`) for visualization, and matplotlib's pyplot (`plt`) for visualization.

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.graphics.api as smg
import matplotlib.pyplot as plt
```

First, make sure that the data file `data-oil-production.csv` is accessible to your Python environment. If you are using your local machine, you can put it in the same directory as this Jupyter notebook. If you are using a cloud service like Google Colab, make sure to upload the file to the remote machine (expand folder icon on left side and click upload button).

Next, read the data file using pandas.

```python
df = pd.read_csv('data-oil-production.csv')
display(df)
```

Statsmodels provides a correlation plot as a diagnostic tool that can help identify highly correlated variables. Note that many of the variables are highly correlated with each other.

```python
fig = smg.plot_corr(df.corr(), xnames=df.columns)
```

Pandas provides a scatter matrix as a diagnostic tool for the same purpose.

```python
figs = pd.plotting.scatter_matrix(df, figsize=(12,12))
```

Next, build an ordinary least squares (OLS) model using statsmodels. A backwards-stepsize procedure starts with an equation to model the `CrudeProduction` variable as a function of the variables `EnergyConsumption`, `NuclearElectricity`, `CoalProduction`, `GasProduction`, `AutoMPG`. Note that the results include a warning of strong multicollinearity.

```python
model = smf.ols('CrudeProduction ~ EnergyConsumption + NuclearElectricity + CoalProduction + GasProduction + AutoMPG', data=df).fit()
print(model.summary())
```

Statsmodels can produce a Q-Q plot (also known as a probability plot) to verify normality of residuals. There are also some options for plotting residuals, but the existing library functions cover more complex topics than what we have discussed in class. The example below plots the predicted crude production versus the model residuals.

```python
fig = sm.qqplot(model.resid, line='s')

plt.figure()
plt.scatter(model.fittedvalues, model.resid)
plt.ylabel('Residuals')
plt.xlabel('Predicted Crude Production')
plt.show()
```
